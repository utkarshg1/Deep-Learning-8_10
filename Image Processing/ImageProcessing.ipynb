# Image Processing 

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Reading colored image in python

```python
import cv2
```

```python
img = cv2.imread('CuteCat.jpeg')
img
```

```python
img.shape
```

### Viewing above array as image

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_rgb
```

```python
plt.imshow(img_rgb)
```

### Reading image as greyscale

```python
img_gray = cv2.imread('CuteCat.jpeg', cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

### Split image to get RGB components

```python
b, g, r = cv2.split(img)
```

```python
b
```

```python
g
```

```python
r
```

```python
plt.imshow(b, cmap='Blues')
```

```python
plt.imshow(g, cmap='Greens')
```

```python
plt.imshow(r, cmap='Reds')
```

### Plotting image with pixel values

```python
a = cv2.imread("Alphabet Example.jpg", cv2.IMREAD_GRAYSCALE)
a
```

```python
a.shape
```

```python
plt.imshow(a, cmap='gray')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(16, 16))
sns.heatmap(a, fmt='d', annot=True)
plt.show()
```
