# Binary Classification with Keras

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Read the train dataset

```python
import pandas as pd
df = pd.read_csv('train_smoker.csv')
df.head()
```

```python
df.info()
```

### Check missing values

```python
df.isna().sum()
```

### Seperate X and Y(Smoking)

```python
X = df.drop(columns=['id', 'smoking'])
Y = df[['smoking']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Preprocessing X with pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

### Apply Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.shape[1]
```

### Create a neural network with keras

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

### Dropout Layer acts as a Regularization. Randomly drop some percentage of neurons

```python
model = Sequential()

# Hidden Layer 1
model.add(Dense(64, activation='relu', input_dim=xtrain.shape[1]))
# Dropout Layer 1
model.add(Dropout(0.2))

# Hidden Layer 2
model.add(Dense(32, activation='relu'))
# Dropout Layer 2
model.add(Dropout(0.2))

# Output Layer
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

### Compile the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'AUC'])
```

### Fit the model

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=60)
```

### Plot the learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'], label='Train BCE')
plt.plot(hist.history['val_loss'], label='Test BCE')
plt.xlabel('epochs')
plt.ylabel('Binary Cross Entropy')
plt.title('Learning Curve')
plt.show()
```

### Evaluate model on train and test

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

### Predicting train and test

```python
yprob_tr = model.predict(xtrain)
yprob_ts = model.predict(xtest)
```

```python
yprob_tr[0:5]
```

### Predicting the results

threshold = 0.5

if prob>=0.5 then smoker else non_smoker


```python
ypred_tr = [1 if i>=0.5 else 0 for i in yprob_tr]
ypred_ts = [1 if i>=0.5 else 0 for i in yprob_ts]
```

```python
ypred_tr[0:5]
```

```python
ypred_ts[0:5]
```

### Create confusion matrix for test results

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

### Classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```
